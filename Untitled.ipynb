```python
data = [
    (2.6, 1),
    (3.8, 1.5),
    (5, 2.1),
    (1.6, 0.5)
]
input_volts, output_volts = list(zip(*data))
```

```python
import pandas as pd
import seaborn as sns
sns.set_theme()
```

```python
df = pd.DataFrame({"Input (V)": input_volts, "Recitfied Output (V)": output_volts})
```

```python
sns.scatterplot(x="Input (V)", y="Recitfied Output (V)", data=df)
```
